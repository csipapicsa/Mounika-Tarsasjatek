```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout, GRU
from tensorflow.keras.optimizers import Adam
```

```python
import scikit-learn
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import lottoMainFunctions as mf
```

```python
tip =  "hatos", 6
df, dfRaw = m.getLotto(m.getLottoDB(tip[0]), type=tip[0])
Years = 10
# cut
dfRaw = dfRaw[0:(Years*52)]
```

# One Hot Encoder

```python
one_hot_encoded = tf.one_hot(dfRaw, depth=45)
transformed_array = tf.reduce_max(one_hot_encoded, axis=1)
transformed_array[1+14:1+14+1,0:6]
```

## Prepare training

```python
window_length = 14 # weeks
# Balls counts
number_of_features = tip[1] # how many ball, 5, 6 
number_of_rows = transformed_array.shape[0]
```

```python
X = np.empty([number_of_rows - window_length, window_length, 45], dtype=float)
y = np.empty([number_of_rows - window_length, 45], dtype=float)

for i in range(0, number_of_rows - window_length):
    #X[i] = transformed_df.iloc[i : i + window_length, 0 : number_of_features].values
    #y[i] = transformed_df.iloc[i + window_length : i + window_length + 1, 0 : number_of_features].values
    X[i] = transformed_array[i : i + window_length, 0 : 45]
    y[i] = transformed_array[i + window_length : i + window_length + 1, 0 : 45]
```

```python
y
```

```python
BATCH_SIZE = 100 # 100
biDirectionalNeurons = 45*3 # 
EPOCHS = 100 # 100

model = Sequential()
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = True)))
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = False)))
#model.add(Dense(90))
model.add(Dense(45,activation='sigmoid'))

def focal_loss(gamma=2.0, alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        pt = y_pred * y_true + (1 - y_pred) * (1 - y_true)
        focal_loss = -alpha * (1 - pt)**gamma * tf.math.log(pt + 1e-7)
        return tf.reduce_mean(focal_loss)
    return focal_loss_fixed


#model.compile(optimizer=Adam(learning_rate=0.0001), loss ='categorical_crossentropy', metrics=['accuracy'])
# Compile the model with focal loss
model.compile(optimizer=Adam(learning_rate=0.0001), loss=focal_loss(), metrics=['accuracy'])

# history = model.fit(x=X, y=y, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=2)

# Calculate class weights
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight('balanced', classes=[0, 1], y=y.flatten())

# Fit the model with class weights
history = model.fit(x=X, y=y, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=2, class_weight=dict(enumerate(class_weights)))
```

```python
BATCH_SIZE = 100 # 100
biDirectionalNeurons = 45*3 # 
EPOCHS = 100 # 100

model = Sequential()

model.add(Bidirectional(GRU(4, return_sequences=True), input_shape=(window_length, 45)))

model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = True)))
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = False)))
#model.add(Dense(90))
model.add(Dense(45,activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.0001), loss ='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(x=X, y=y, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=2)
```

```python
BATCH_SIZE = 100 # 100
biDirectionalNeurons = 45 # 
EPOCHS = 500 # 100

model = Sequential()

model.add(Bidirectional(GRU(4, return_sequences=True), input_shape=(window_length, 45)))
#model.add(Bidirectional(GRU(45, return_sequences=True), input_shape=(window_length, 45)))
#model.add(Bidirectional(GRU(4, return_sequences=True), input_shape=(window_length, 45)))
model.add(Bidirectional(LSTM(biDirectionalNeurons, input_shape = (window_length, 45), return_sequences = False)))
model.add(Dense(45,activation='sigmoid'))

model.compile(optimizer=Adam(learning_rate=0.0001), loss ='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(x=X, y=y, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=2)
```

```python
training_loss = history.history['loss']

# Plot the training loss
plt.plot(range(1, len(training_loss) + 1), training_loss, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

training_accuracy = history.history['accuracy']
plt.plot(range(1, len(training_accuracy) + 1), training_accuracy, label='Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
